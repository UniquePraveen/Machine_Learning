```python
x = data.iloc[:,1:5]
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
x = scaler.fit_transform(x)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y ,test_size=0.2)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=1)
model.fit(x_train,y_train)
    
```

```python
pred = model.predict(x_test)
pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test,pred)
accuracy
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,pred)
cm
```

```python
result = pd.DataFrame(data=[y_test.values,pred], index = ['y_test','pred'])
result.transpose()
```

```python
correct_sum =[ ]
for i in range(1,20):
    model = KNeighborsClassifier(n_neighbors=i)
    model.fit(x_train,y_train)
    pred = model.predict(x_test)
    correct = np.sum(pred == y_test) 
    correct_sum.append(correct)
```

```python
correct_sum
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(range(1, 20), correct_sum, marker='o', linestyle='-', color='r')
plt.title('Correct Predictions vs. Number of Neighbors')
plt.xlabel('Number of Neighbors (k)')
plt.ylabel('Correct Predictions')
plt.show()
```

```python
result = pd.DataFrame(data=correct_sum)
result.index = result.index + 1
result.T
```
